```python
import fitz

# Open the PDF document
doc = fitz.open("ML 2021390_Hardik Singh.docx-7.pdf")

text = ""

# Loop through all pages of the document
for page in doc:
    # Extract text from the current page and concatenate it to the existing text
    text += page.get_text()

# Close the document
doc.close()
```

```python
text=text.lower()
text = text.split('\n')
text
```

```python
keywords = ["intro", "education", "skills", "internship", "projects", 'positions of responsibility', "awards and achievements"]
sections = {}
sec=[]
i=0
for line in text:
    if(line in keywords):
        
        sections[keywords[i]] = sec
        i+=1
        sec=[]
    else:
        sec.append(line)
sections[keywords[i]]=sec
```
